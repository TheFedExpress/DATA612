## YouTube Recommendation Engine

The Google subsidiary has an estimated value of over $100 billion, around 70 percent of Netflix, so it is no surprise the site's recommendation engine is among the more advanced in the marketplace.  Gone are the days when matrix factorization alone is powerful enough to drive a cutting edge engine.  YouTube now uses a multistage deep learning model to feed content to its users.

This diagram from the "Towards Data Science" link below summarized the network architecture.



The two primary stages are candidate generation and ranking.  

### Candidate Generation

The goal of this stage is to pare down the list of possible videos to the "candidates" for ranking.  This stage uses a cousin of matrix factorization, embeddings.  Just like factorization, matrices of users, videos, and ratings can be fed into the embedding layer and projected onto a low-dimensional space.  However, this is a non-linear projection using the y-labels to generate the feature space.  The generated features are much more predictive than PCA or SVD because it is supervised learning techique.  The non-linear aspect is just icing on the cake.  The candidate generation stage also uses a few other features, such as the number of views to help widdle the list down.  I would think this feature carries a lot of weight because I rarely see videos with less than 50,000 views recommended.

### Ranking

Once the candiates are selected, the engine needs to rank them for the user.  At this stage, richer features related to the video's content are added.  One of the more interesting features is the recency of the video, when it was posted.  Older videos naturally would have more views than new videos, which would affect the videos embeddings, the users embeddings, and the "views" feature.  Recency helps to compensate for this and keeps the content on youtube fresh and effective.  The other interesting feature, not related to similarity, is "recent views".  This allows the problem to be framed as timeseries prediction, which makes perfect sense if the goal is to predict what the user wants to see at that moment.

### Personal Experience

I am sometimes annoyed by the fact that YouTube doesn't quite "read my mind" like Netflix does.  My main gripe is that when I watching learning-related videos, I want to see a sequence, not similar content and not novel conent.  I could see a mix of similar and sequence-related in case you want to re-enfore instead of moving onto the next topic, but it doesn't seem to get the mix right.

To give credit to YouTube, they are dealing with incredible amounts of data so encoding too many features would render the service too slow.  They're also dealing with far more videos than Netflix is movies.  

## Attacks on Recommender Systems

In the mid-2000s, I relied heavily on IMDB ratings so when the movie "Snakes on a Plane" had a rating close to that of "The Godfather", so this naturally piqued my interst.  Given the strange title and campy preview I was a little skeptical so I consulted some critic's reviews before trying to get tickets, and sure enough, they were pretty bad.  I'm still not certain this was a coordinated effort, but the reviews were certainly not legitimate (to confirm my memory hadn't failed me, I found the IMDB reviews below, which mention the unusually high rating).

Whether these ratings were the result of an internet meme or something more sinister doesn't change the fact that IMDB would rather not have their ratings compromised.  The site can (and probably does) take some steps to ensure ratings come from legitimate users by removing ratings from users with traits like: 

- Users with too many ratings
- Users with too few ratings
- Users with mimimal suspect account information
- Users logging in from many different IPs in a short amount of time

When legitimate users coordinate an attack, it can be a little more difficult to stop.  IMDB could look at movies with:

- inordinate counts of 1s and 10s for ratings
- spikes in the volume of ratings
- extreme clumps of ratings within certain demographics

Markers like these could flag suspcious activity for further investigation.  If the activity were determined to be malicious, some of the ratings could be removed.  This isn't perfect, but it could at least curb abuse.
